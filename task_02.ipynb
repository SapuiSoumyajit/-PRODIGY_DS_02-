```python
import pandas as pd

df = pd.read_csv("train.csv")

# Impute Age
df['Age'].fillna(df['Age'].median(), inplace=True)

# Impute Embarked
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop Cabin
df.drop(columns=['Cabin'], inplace=True)
```

```python
df['Survived'].value_counts().plot(kind='bar', title='Survival Count', color=['salmon', 'skyblue'])
```

```python
import seaborn as sns
sns.barplot(x='Sex', y='Survived', data=df)
```

```python
sns.barplot(x='Pclass', y='Survived', data=df)
```

```python
sns.histplot(df[df['Survived']==1]['Age'], color='green', label='Survived', kde=True)
sns.histplot(df[df['Survived']==0]['Age'], color='red', label='Not Survived', kde=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
